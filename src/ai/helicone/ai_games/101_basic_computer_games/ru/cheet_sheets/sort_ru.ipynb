# –ê–ª–≥–æ—Ä–∏—Ç–º—ã —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏



*   –ü–µ—Ä–≤—ã–π —ç–ª–µ–º–µ–Ω—Ç ‚Äì —ç—Ç–æ —Ä–∞–∑–º–µ—Ä —Ñ—Ä—É–∫—Ç–∞:
    *   üçé (–º–µ–ª–∫–∏–π) ‚Äì —è–±–ª–æ–∫–æ
    *   üçê (—Å—Ä–µ–¥–Ω–∏–π) ‚Äì –≥—Ä—É—à–∞
    *   üçâ (–±–æ–ª—å—à–æ–π) ‚Äì –¥—ã–Ω—è
    *   üß∫ (–æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π) ‚Äì –∫–æ—Ä–∑–∏–Ω–∞
*   –í—Ç–æ—Ä–æ–π —ç–ª–µ–º–µ–Ω—Ç ‚Äì —ç—Ç–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã–π –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä, –¥–ª—è —Ä–∞–±–æ—Ç—ã –ø—Ä–æ–≥—Ä–∞–º–º—ã.

–ü—Ä–∏–º–µ—Ä: `(üçé, 1)` ‚Äì —ç—Ç–æ –º–µ–ª–∫–æ–µ —è–±–ª–æ–∫–æ —Å –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä–æ–º 1.

**–ê–ª–≥–æ—Ä–∏—Ç–º—ã —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ (—Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ —Ä–∞–∑–º–µ—Ä—É —Ñ—Ä—É–∫—Ç–∞):**

**–ê–ª–≥–æ—Ä–∏—Ç–º—ã —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ (—Å—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ —Ä–∞–∑–º–µ—Ä—É —Ñ—Ä—É–∫—Ç–∞):**

1.  **–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø—É–∑—ã—Ä—å–∫–æ–º (Bubble Sort):**
    *   –ê–ª–≥–æ—Ä–∏—Ç–º —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç —Å–æ—Å–µ–¥–Ω–∏–µ —Ñ—Ä—É–∫—Ç—ã –ø–æ —Ä–∞–∑–º–µ—Ä—É. –ï—Å–ª–∏ —Ñ—Ä—É–∫—Ç –±–æ–ª—å—à–µ, —á–µ–º —Å–æ—Å–µ–¥–Ω–∏–π, –æ–Ω –º–µ–Ω—è–µ—Ç—Å—è —Å –Ω–∏–º –º–µ—Å—Ç–∞–º–∏.
    *   –≠—Ç–æ—Ç –ø—Ä–æ—Ü–µ—Å—Å –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –≤–µ—Å—å —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –Ω–µ –±—É–¥–µ—Ç –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω –æ—Ç –º–µ–Ω—å—à–µ–≥–æ –∫ –±–æ–ª—å—à–µ–º—É.
    *   –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø—É–∑—ã—Ä—å–∫–æ–º –ø—Ä–æ—Å—Ç–∞, –Ω–æ –Ω–µ –ø–æ–¥—Ö–æ–¥–∏—Ç –¥–ª—è –±–æ–ª—å—à–∏—Ö —Å–ø–∏—Å–∫–æ–≤ —Ñ—Ä—É–∫—Ç–æ–≤, —Ç–∞–∫ –∫–∞–∫ –Ω–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞.

2.  **–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏ (Insertion Sort):**
    *   –ê–ª–≥–æ—Ä–∏—Ç–º —Å—Ç—Ä–æ–∏—Ç –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫, –¥–æ–±–∞–≤–ª—è—è –≤ –Ω–µ–≥–æ —Ñ—Ä—É–∫—Ç—ã –æ–¥–∏–Ω –∑–∞ –¥—Ä—É–≥–∏–º. –ù–æ–≤—ã–π —Ñ—Ä—É–∫—Ç –≤—Å—Ç–∞–≤–ª—è–µ—Ç—Å—è –≤ –Ω—É–∂–Ω–æ–µ –º–µ—Å—Ç–æ, —á—Ç–æ–±—ã —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –ø–æ—Ä—è–¥–æ–∫ –ø–æ —Ä–∞–∑–º–µ—Ä—É.
    *   –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏ —Ö–æ—Ä–æ—à–∞ –¥–ª—è –Ω–µ–±–æ–ª—å—à–∏—Ö —Å–ø–∏—Å–∫–æ–≤ –∏–ª–∏ –¥–ª—è —Ç–µ—Ö, –≥–¥–µ –¥–∞–Ω–Ω—ã–µ —É–∂–µ –ø–æ—á—Ç–∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã.

3.  **–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º (Selection Sort):**
    *   –ê–ª–≥–æ—Ä–∏—Ç–º –Ω–∞—Ö–æ–¥–∏—Ç —Å–∞–º—ã–π –º–∞–ª–µ–Ω—å–∫–∏–π —Ñ—Ä—É–∫—Ç –≤ –Ω–µ–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏ —Å–ø–∏—Å–∫–∞. –ó–∞—Ç–µ–º –æ–Ω —Å—Ç–∞–≤–∏—Ç —ç—Ç–æ—Ç —Ñ—Ä—É–∫—Ç –Ω–∞ –ø–µ—Ä–≤–æ–µ –º–µ—Å—Ç–æ –≤ –Ω–µ–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏ —Å–ø–∏—Å–∫–∞.
    *   –≠—Ç–æ—Ç –ø—Ä–æ—Ü–µ—Å—Å –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è –¥–æ —Ç–µ—Ö –ø–æ—Ä, –ø–æ–∫–∞ –≤—Å–µ —Ñ—Ä—É–∫—Ç—ã –Ω–µ –±—É–¥—É—Ç –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω—ã.
    *   –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º –ø—Ä–æ—Å—Ç–∞, –Ω–æ –Ω–µ—ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–∞ –¥–ª—è –±–æ–ª—å—à–∏—Ö —Å–ø–∏—Å–∫–æ–≤.


```python
from typing import List, Tuple

def compare_fruits(fruit1: Tuple[str, int], fruit2: Tuple[str, int]) -> int:
    """
    –°—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç –¥–≤–∞ —Ñ—Ä—É–∫—Ç–∞ –ø–æ —Ä–∞–∑–º–µ—Ä—É.

    Args:
        fruit1: –ö–æ—Ä—Ç–µ–∂ (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).
        fruit2: –ö–æ—Ä—Ç–µ–∂ (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        -1, –µ—Å–ª–∏ fruit1 –º–µ–Ω—å—à–µ fruit2, 1, –µ—Å–ª–∏ fruit1 –±–æ–ª—å—à–µ fruit2, 0, –µ—Å–ª–∏ —Ä–∞–≤–Ω—ã.
    """
    order = {"üçé": 0, "üçê": 1, "üçâ": 2, "üß∫": 3}  # –û–ø—Ä–µ–¥–µ–ª—è—é –ø–æ—Ä—è–¥–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –ø–æ —Ä–∞–∑–º–µ—Ä—É
    size1 = order.get(fruit1[0]) # –ü–æ–ª—É—á–∞—é —Ä–∞–∑–º–µ—Ä –ø–µ—Ä–≤–æ–≥–æ —Ñ—Ä—É–∫—Ç–∞
    size2 = order.get(fruit2[0]) # –ü–æ–ª—É—á–∞—é —Ä–∞–∑–º–µ—Ä –≤—Ç–æ—Ä–æ–≥–æ —Ñ—Ä—É–∫—Ç–∞
    if size1 < size2: # –ï—Å–ª–∏ —Ä–∞–∑–º–µ—Ä –ø–µ—Ä–≤–æ–≥–æ —Ñ—Ä—É–∫—Ç–∞ –º–µ–Ω—å—à–µ, –≤–æ–∑–≤—Ä–∞—â–∞—é -1
        return -1
    elif size1 > size2: # –ï—Å–ª–∏ —Ä–∞–∑–º–µ—Ä –ø–µ—Ä–≤–æ–≥–æ —Ñ—Ä—É–∫—Ç–∞ –±–æ–ª—å—à–µ, –≤–æ–∑–≤—Ä–∞—â–∞—é 1
        return 1
    else: # –ï—Å–ª–∏ —Ä–∞–∑–º–µ—Ä—ã —Ä–∞–≤–Ω—ã, –≤–æ–∑–≤—Ä–∞—â–∞—é 0
      return 0

def bubble_sort(fruits: List[Tuple[str, int]]) -> List[Tuple[str, int]]:
    """
    –°–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –ø–æ —Ä–∞–∑–º–µ—Ä—É, –∏—Å–ø–æ–ª—å–∑—É—è –∞–ª–≥–æ—Ä–∏—Ç–º "–ø—É–∑—ã—Ä—å–∫–æ–≤–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞".

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π.
    """
    n = len(fruits)  # –ü–æ–ª—É—á–∞—é –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ—Ä—É–∫—Ç–æ–≤
    for i in range(n):  # –ü—Ä–æ—Ö–æ–∂—É –ø–æ —Å–ø–∏—Å–∫—É n —Ä–∞–∑
        for j in range(0, n - i - 1):  # –ü—Ä–æ—Ö–æ–∂—É –ø–æ –Ω–µ–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏ —Å–ø–∏—Å–∫–∞
            if compare_fruits(fruits[j], fruits[j + 1]) == 1:  # –ï—Å–ª–∏ —Ñ—Ä—É–∫—Ç —Å–ª–µ–≤–∞ –±–æ–ª—å—à–µ, —á–µ–º —Ñ—Ä—É–∫—Ç —Å–ø—Ä–∞–≤–∞
                fruits[j], fruits[j + 1] = fruits[j + 1], fruits[j]  # –ú–µ–Ω—è—é –º–µ—Å—Ç–∞–º–∏
    return fruits


def insertion_sort(fruits: List[Tuple[str, int]]) -> List[Tuple[str, int]]:
    """
    –°–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –ø–æ —Ä–∞–∑–º–µ—Ä—É, –∏—Å–ø–æ–ª—å–∑—É—è –∞–ª–≥–æ—Ä–∏—Ç–º "—Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏".

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π.
    """
    for i in range(1, len(fruits)): # –ù–∞—á–∏–Ω–∞—é —Å–æ –≤—Ç–æ—Ä–æ–≥–æ —Ñ—Ä—É–∫—Ç–∞ (–ø–µ—Ä–≤—ã–π —Å—á–∏—Ç–∞–µ—Ç—Å—è –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–º)
        key = fruits[i] # –ë–µ—Ä—É —Å–ª–µ–¥—É—é—â–∏–π —Ñ—Ä—É–∫—Ç
        j = i - 1 # –ò–Ω–¥–µ–∫—Å –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ —Ñ—Ä—É–∫—Ç–∞
        while j >= 0 and compare_fruits(fruits[j], key) == 1: # –ò—â—É –ø–æ–∑–∏—Ü–∏—é –≤ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏, –∫—É–¥–∞ –≤—Å—Ç–∞–≤–∏—Ç—å —Ñ—Ä—É–∫—Ç
            fruits[j + 1] = fruits[j] # –°–¥–≤–∏–≥–∞—é —Ñ—Ä—É–∫—Ç—ã, —á—Ç–æ –±—ã –æ—Å–≤–æ–±–æ–¥–∏—Ç—å –º–µ—Å—Ç–æ –¥–ª—è –Ω–æ–≤–æ–≥–æ
            j -= 1
        fruits[j + 1] = key # –í—Å—Ç–∞–≤–ª—è—é —Ñ—Ä—É–∫—Ç –Ω–∞ –Ω—É–∂–Ω–æ–µ –º–µ—Å—Ç–æ
    return fruits

def selection_sort(fruits: List[Tuple[str, int]]) -> List[Tuple[str, int]]:
    """
    –°–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –ø–æ —Ä–∞–∑–º–µ—Ä—É, –∏—Å–ø–æ–ª—å–∑—É—è –∞–ª–≥–æ—Ä–∏—Ç–º "—Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º".

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π.
    """
    n = len(fruits) # –ü–æ–ª—É—á–∞—é –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ñ—Ä—É–∫—Ç–æ–≤ –≤ —Å–ø–∏—Å–∫–µ
    for i in range(n): # –ü—Ä–æ—Ö–æ–∂—É –ø–æ –≤—Å–µ–º —Ñ—Ä—É–∫—Ç–∞–º –≤ —Å–ø–∏—Å–∫–µ
        min_index = i # –ò–Ω–¥–µ–∫—Å —Å–∞–º–æ–≥–æ –º–∞–ª–µ–Ω—å–∫–æ–≥–æ —Ñ—Ä—É–∫—Ç–∞
        for j in range(i + 1, n): # –ò—â—É —Å–∞–º—ã–π –º–∞–ª–µ–Ω—å–∫–∏–π —Ñ—Ä—É–∫—Ç –≤ –Ω–µ–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏
            if compare_fruits(fruits[j], fruits[min_index]) == -1: # –ï—Å–ª–∏ –Ω–∞—à–µ–ª —Ñ—Ä—É–∫—Ç –º–µ–Ω—å—à–µ —Ç–µ–∫—É—â–µ–≥–æ –º–∏–Ω–∏–º—É–º–∞
                min_index = j # –ó–∞–ø–æ–º–∏–Ω–∞—é –∏–Ω–¥–µ–∫—Å –Ω–æ–≤–æ–≥–æ –º–∏–Ω–∏–º—É–º–∞
        fruits[i], fruits[min_index] = fruits[min_index], fruits[i] # –ú–µ–Ω—è—é —Ç–µ–∫—É—â–∏–π —Ñ—Ä—É–∫—Ç —Å —Å–∞–º—ã–º –º–∞–ª–µ–Ω—å–∫–∏–º –∏–∑ –Ω–µ–æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏
    return fruits

def display_fruits(fruits: List[Tuple[str, int]]) -> str:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –≤ —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è.

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –°—Ç—Ä–æ–∫–∞ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ —Ñ—Ä—É–∫—Ç–æ–≤.
    """
    return ", ".join(f"{fruit[0]}{fruit[1]}" for fruit in fruits)  # –°–æ–±–∏—Ä–∞—é —Å—Ç—Ä–æ–∫—É –¥–ª—è –≤—ã–≤–æ–¥–∞


# –°–æ–∑–¥–∞—é —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
fruits = [
    ("üçâ", 1), ("üçé", 2), ("üçê", 3), ("üß∫", 4), ("üçé", 5), ("üçâ", 6), ("üçê", 7),
    ("üçé", 8), ("üß∫", 9), ("üçâ", 10), ("üçê", 11), ("üçé", 12)
]

print("–ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤: " + display_fruits(fruits))  # –í—ã–≤–æ–∂—É –∏—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫
print("–ü—Ä–∏–º–µ—Ä—ã: –Ø–±–ª–æ–∫–æ (üçé) < –ì—Ä—É—à–∏ (üçê) < –î—ã–Ω—è (üçâ) < –ö–æ—Ä–∑–∏–Ω—ã (üß∫)")  # –ü–æ–∫–∞–∑—ã–≤–∞—é –ø–æ—Ä—è–¥–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤

# –°–æ—Ä—Ç–∏—Ä—É—é –ø—É–∑—ã—Ä—å–∫–æ–º
sorted_fruits_bubble = bubble_sort(fruits.copy()) # –°–æ—Ä—Ç–∏—Ä—É—é –∫–æ–ø–∏—é —Å–ø–∏—Å–∫–∞
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø—É–∑—ã—Ä—å–∫–æ–º: " + display_fruits(sorted_fruits_bubble)) # –í—ã–≤–æ–∂—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç

# –°–æ—Ä—Ç–∏—Ä—É—é –≤—Å—Ç–∞–≤–∫–∞–º–∏
sorted_fruits_insertion = insertion_sort(fruits.copy()) # –°–æ—Ä—Ç–∏—Ä—É—é –∫–æ–ø–∏—é —Å–ø–∏—Å–∫–∞
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏: " + display_fruits(sorted_fruits_insertion)) # –í—ã–≤–æ–∂—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç

# –°–æ—Ä—Ç–∏—Ä—É—é –≤—ã–±–æ—Ä–æ–º
sorted_fruits_selection = selection_sort(fruits.copy()) # –°–æ—Ä—Ç–∏—Ä—É—é –∫–æ–ø–∏—é —Å–ø–∏—Å–∫–∞
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º: " + display_fruits(sorted_fruits_selection)) # –í—ã–≤–æ–∂—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç
```

**–†–∞–∑—ä—è—Å–Ω–µ–Ω–∏–µ –∫–æ–¥–∞:**

1.  **`compare_fruits(fruit1, fruit2)`:** –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è —Å—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç –¥–≤–∞ —Ñ—Ä—É–∫—Ç–∞ –ø–æ —Ä–∞–∑–º–µ—Ä—É –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç -1, –µ—Å–ª–∏ –ø–µ—Ä–≤—ã–π —Ñ—Ä—É–∫—Ç –º–µ–Ω—å—à–µ, 1, –µ—Å–ª–∏ –±–æ–ª—å—à–µ, –∏ 0, –µ—Å–ª–∏ –æ–Ω–∏ —Ä–∞–≤–Ω—ã. –Ø –∏—Å–ø–æ–ª—å–∑—É—é —Å–ª–æ–≤–∞—Ä—å `order` –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ—Ä—è–¥–∫–∞ —Ä–∞–∑–º–µ—Ä–æ–≤ —Ñ—Ä—É–∫—Ç–æ–≤.
2.  **`bubble_sort(fruits)`:** –Ø —Ä–µ–∞–ª–∏–∑—É—é –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –ø—É–∑—ã—Ä—å–∫–æ–º, –≥–¥–µ —Å–æ—Å–µ–¥–Ω–∏–µ —Ñ—Ä—É–∫—Ç—ã —Å—Ä–∞–≤–Ω–∏–≤–∞—é—Ç—Å—è –∏ –º–µ–Ω—è—é—Ç—Å—è –º–µ—Å—Ç–∞–º–∏, –µ—Å–ª–∏ –æ–Ω–∏ –≤ –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–º –ø–æ—Ä—è–¥–∫–µ.
3.  **`insertion_sort(fruits)`:** –Ø —Ä–µ–∞–ª–∏–∑—É—é –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—Å—Ç–∞–≤–∫–∞–º–∏, –≥–¥–µ –∫–∞–∂–¥—ã–π –Ω–æ–≤—ã–π —Ñ—Ä—É–∫—Ç –≤—Å—Ç–∞–≤–ª—è–µ—Ç—Å—è –≤ –Ω—É–∂–Ω–æ–µ –º–µ—Å—Ç–æ –≤ —É–∂–µ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —á–∞—Å—Ç–∏ —Å–ø–∏—Å–∫–∞.
4.  **`selection_sort(fruits)`:** –Ø —Ä–µ–∞–ª–∏–∑—É—é –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –≤—ã–±–æ—Ä–æ–º, –≥–¥–µ –≤ –∫–∞–∂–¥–æ–º –ø—Ä–æ—Ö–æ–¥–µ —è –Ω–∞—Ö–æ–∂—É –Ω–∞–∏–º–µ–Ω—å—à–∏–π —Ñ—Ä—É–∫—Ç –∏ —Å—Ç–∞–≤–ª—é –µ–≥–æ –Ω–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ–µ –º–µ—Å—Ç–æ.
5.  **`display_fruits(fruits)`:** –≠—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –≤ —Å—Ç—Ä–æ–∫—É –¥–ª—è —É–¥–æ–±–Ω–æ–≥–æ –≤—ã–≤–æ–¥–∞.
6.  **–ü—Ä–∏–º–µ—Ä—ã:** –í –∫–æ–Ω—Ü–µ —è —Å–æ–∑–¥–∞—é —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –∏ –ø—Ä–∏–º–µ–Ω—è—é –∫ –Ω–µ–º—É –≤—Å–µ —Ç—Ä–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏, –≤—ã–≤–æ–¥—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∫–∞–∂–¥–æ–≥–æ –∏–∑ –Ω–∏—Ö. –Ø —Ç–∞–∫–∂–µ –ø–æ–∫–∞–∑—ã–≤–∞—é —Ç–µ–±–µ –ø–æ—Ä—è–¥–æ–∫, –≤ –∫–æ—Ç–æ—Ä–æ–º —Ñ—Ä—É–∫—Ç—ã —Å–æ—Ä—Ç–∏—Ä—É—é—Ç—Å—è.


````python
from typing import List, Tuple

def compare_fruits(fruit1: Tuple[str, int], fruit2: Tuple[str, int]) -> int:
    """
    –°—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç –¥–≤–∞ —Ñ—Ä—É–∫—Ç–∞ –ø–æ —Ä–∞–∑–º–µ—Ä—É.

    Args:
        fruit1: –ö–æ—Ä—Ç–µ–∂ (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).
        fruit2: –ö–æ—Ä—Ç–µ–∂ (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        -1, –µ—Å–ª–∏ fruit1 –º–µ–Ω—å—à–µ fruit2, 1, –µ—Å–ª–∏ fruit1 –±–æ–ª—å—à–µ fruit2, 0, –µ—Å–ª–∏ —Ä–∞–≤–Ω—ã.
    """
    order = {"üçé": 0, "üçê": 1, "üçâ": 2, "üß∫": 3}
    size1 = order.get(fruit1[0])
    size2 = order.get(fruit2[0])
    if size1 < size2:
        return -1
    elif size1 > size2:
        return 1
    else:
      return 0

def bubble_sort(fruits: List[Tuple[str, int]]) -> List[Tuple[str, int]]:
    """
    –°–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –ø–æ —Ä–∞–∑–º–µ—Ä—É, –∏—Å–ø–æ–ª—å–∑—É—è –∞–ª–≥–æ—Ä–∏—Ç–º "–ø—É–∑—ã—Ä—å–∫–æ–≤–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞".

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π.
    """
    n = len(fruits)
    for i in range(n):
        for j in range(0, n - i - 1):
            if compare_fruits(fruits[j], fruits[j + 1]) == 1:
                fruits[j], fruits[j + 1] = fruits[j + 1], fruits[j]
    return fruits


def insertion_sort(fruits: List[Tuple[str, int]]) -> List[Tuple[str, int]]:
    """
    –°–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –ø–æ —Ä–∞–∑–º–µ—Ä—É, –∏—Å–ø–æ–ª—å–∑—É—è –∞–ª–≥–æ—Ä–∏—Ç–º "—Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏".

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π.
    """
    for i in range(1, len(fruits)):
        key = fruits[i]
        j = i - 1
        while j >= 0 and compare_fruits(fruits[j], key) == 1:
            fruits[j + 1] = fruits[j]
            j -= 1
        fruits[j + 1] = key
    return fruits

def selection_sort(fruits: List[Tuple[str, int]]) -> List[Tuple[str, int]]:
    """
    –°–æ—Ä—Ç–∏—Ä—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –ø–æ —Ä–∞–∑–º–µ—Ä—É, –∏—Å–ø–æ–ª—å–∑—É—è –∞–ª–≥–æ—Ä–∏—Ç–º "—Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º".

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –û—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π.
    """
    n = len(fruits)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if compare_fruits(fruits[j], fruits[min_index]) == -1:
                min_index = j
        fruits[i], fruits[min_index] = fruits[min_index], fruits[i]
    return fruits

def display_fruits(fruits: List[Tuple[str, int]]) -> str:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –≤ —Å—Ç—Ä–æ–∫—É –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è.

    Args:
        fruits: –°–ø–∏—Å–æ–∫ –∫–æ—Ä—Ç–µ–∂–µ–π (—Ä–∞–∑–º–µ—Ä, –∏–¥–µ–Ω—Ç–∏—Ñ–∏–∫–∞—Ç–æ—Ä).

    Returns:
        –°—Ç—Ä–æ–∫–∞ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ —Ñ—Ä—É–∫—Ç–æ–≤.
    """
    return ", ".join(f"{fruit[0]}{fruit[1]}" for fruit in fruits)


# –°–æ–∑–¥–∞—é —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤ –¥–ª—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
fruits = [
    ("üçâ", 1), ("üçé", 2), ("üçê", 3), ("üß∫", 4), ("üçé", 5), ("üçâ", 6), ("üçê", 7),
    ("üçé", 8), ("üß∫", 9), ("üçâ", 10), ("üçê", 11), ("üçé", 12)
]

print("–ò—Å—Ö–æ–¥–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Ñ—Ä—É–∫—Ç–æ–≤: " + display_fruits(fruits))
print("–ü—Ä–∏–º–µ—Ä—ã: –Ø–±–ª–æ–∫–æ (üçé) < –ì—Ä—É—à–∏ (üçê) < –î—ã–Ω—è (üçâ) < –ö–æ—Ä–∑–∏–Ω—ã (üß∫)")

# –°–æ—Ä—Ç–∏—Ä—É—é –ø—É–∑—ã—Ä—å–∫–æ–º
sorted_fruits_bubble = bubble_sort(fruits.copy())
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø—É–∑—ã—Ä—å–∫–æ–º: " + display_fruits(sorted_fruits_bubble))

# –°–æ—Ä—Ç–∏—Ä—É—é –≤—Å—Ç–∞–≤–∫–∞–º–∏
sorted_fruits_insertion = insertion_sort(fruits.copy())
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—Å—Ç–∞–≤–∫–∞–º–∏: " + display_fruits(sorted_fruits_insertion))

# –°–æ—Ä—Ç–∏—Ä—É—é –≤—ã–±–æ—Ä–æ–º
sorted_fruits_selection = selection_sort(fruits.copy())
print("–°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –≤—ã–±–æ—Ä–æ–º: " + display_fruits(sorted_fruits_selection))
````
