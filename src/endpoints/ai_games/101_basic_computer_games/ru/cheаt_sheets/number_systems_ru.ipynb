1.  **–í—ã–±–æ—Ä –æ—Å–Ω–æ–≤–∞–Ω–∏—è:** –í—ã–±–∏—Ä–∞–µ–º —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ `b`, –∫–æ—Ç–æ—Ä–æ–µ –±—É–¥–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–µ–º –Ω–∞—à–µ–π —Å–∏—Å—Ç–µ–º—ã.
2.  **–í—ã–±–æ—Ä —Ü–∏—Ñ—Ä:** –ù–∞–º –Ω—É–∂–Ω–æ `b` —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ü–∏—Ñ—Ä. –û–±—ã—á–Ω–æ —ç—Ç–æ 0, 1, 2, ..., `b-1`.
3.  **–ó–∞–ø–∏—Å—å —á–∏—Å–ª–∞:** –ß–∏—Å–ª–æ –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ü–∏—Ñ—Ä. –ó–Ω–∞—á–µ–Ω–∏–µ –∫–∞–∂–¥–æ–π —Ü–∏—Ñ—Ä—ã —É–º–Ω–æ–∂–∞–µ—Ç—Å—è –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–µ, –≤–æ–∑–≤–µ–¥–µ–Ω–Ω–æ–µ –≤ —Å—Ç–µ–ø–µ–Ω—å, —Ä–∞–≤–Ω—É—é –µ–µ –ø–æ–∑–∏—Ü–∏–∏ (–Ω–∞—á–∏–Ω–∞—è —Å 0 —Å–ø—Ä–∞–≤–∞).

**–§–æ—Ä–º—É–ª–∞ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –∑–Ω–∞—á–µ–Ω–∏—è —á–∏—Å–ª–∞:**

–ï—Å–ª–∏ —É –Ω–∞—Å –µ—Å—Ç—å —á–∏—Å–ª–æ, –∑–∞–ø–∏—Å–∞–Ω–Ω–æ–µ –≤ –≤–∏–¥–µ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Ü–∏—Ñ—Ä `d‚Çô d‚Çô‚Çã‚ÇÅ ... d‚ÇÅ d‚ÇÄ`, —Ç–æ –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –ø–æ —Ñ–æ—Ä–º—É–ª–µ:

`–∑–Ω–∞—á–µ–Ω–∏–µ = d‚Çô * b‚Åø + d‚Çô‚Çã‚ÇÅ * b‚Åø‚Åª¬π + ... + d‚ÇÅ * b¬π + d‚ÇÄ * b‚Å∞`

–ì–¥–µ:

*   `d·µ¢` - —Ü–∏—Ñ—Ä–∞ –≤ i-–æ–º —Ä–∞–∑—Ä—è–¥–µ
*   `b` - –æ—Å–Ω–æ–≤–∞–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã —Å—á–∏—Å–ª–µ–Ω–∏—è
*   `i` - –Ω–æ–º–µ—Ä —Ä–∞–∑—Ä—è–¥–∞ (—Å–ø—Ä–∞–≤–∞ –Ω–∞–ª–µ–≤–æ, –Ω–∞—á–∏–Ω–∞—è —Å 0)

**–ü—Ä–∏–º–µ—Ä:**

`1 * 10¬≤ + 2 * 10¬π + 3 * 10‚Å∞ = 100 + 20 + 3 = 123`

**–ü–æ—Ä—è–¥–∫–∏ —Å—á–∏—Å–ª–µ–Ω–∏—è (—Ä–∞–∑—Ä—è–¥—ã):**

–ü–æ—Ä—è–¥–∫–∏, –∫–∞–∫ –º—ã —É–∂–µ –≥–æ–≤–æ—Ä–∏–ª–∏, —ç—Ç–æ –ø–æ–∑–∏—Ü–∏–∏ —Ü–∏—Ñ—Ä –≤ —á–∏—Å–ª–µ, –∫–∞–∂–¥–∞—è –ø–æ–∑–∏—Ü–∏—è –∏–º–µ–µ—Ç —Å–≤–æ–π –≤–µ—Å, –∫–æ—Ç–æ—Ä—ã–π –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –æ—Å–Ω–æ–≤–∞–Ω–∏–µ–º –≤ —Å—Ç–µ–ø–µ–Ω–∏ –µ–µ –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –Ω–æ–º–µ—Ä–∞.
*   `d‚ÇÄ`: –µ–¥–∏–Ω–∏—Ü—ã (`b‚Å∞`)
*   `d‚ÇÅ`: `b` (`b¬π`)
*   `d‚ÇÇ`: `b¬≤`
*   `d‚ÇÉ`: `b¬≥`
*   –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ

**–ü—Ä–∞–≤–∏–ª–∞:**

1.  **–î–∏–∞–ø–∞–∑–æ–Ω —Ü–∏—Ñ—Ä:** –ò—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è —Ü–∏—Ñ—Ä—ã –æ—Ç 0 –¥–æ `b-1`.
2.  **–ü–æ–∑–∏—Ü–∏–æ–Ω–Ω—ã–π –ø—Ä–∏–Ω—Ü–∏–ø:** –ó–Ω–∞—á–µ–Ω–∏–µ —Ü–∏—Ñ—Ä—ã –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –µ—ë –ø–æ–∑–∏—Ü–∏–∏.
3.  **–ü–µ—Ä–µ—Ö–æ–¥ –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —Ä–∞–∑—Ä—è–¥—É:** –ö–æ–≥–¥–∞ –≤ —Ä–∞–∑—Ä—è–¥–µ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ `b`, –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç –ø–µ—Ä–µ–Ω–æ—Å –≤ —Å–ª–µ–¥—É—é—â–∏–π —Ä–∞–∑—Ä—è–¥ (–∞–Ω–∞–ª–æ–≥ —Ç–æ–≥–æ –∫–∞–∫ –ø–æ—Å–ª–µ 9 –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–π —Å–∏—Å—Ç–µ–º–µ –ø—Ä–∏–±–∞–≤–ª—è–µ—Ç—Å—è 1 –∫ —Å–ª–µ–¥—É—é—â–µ–º—É —Ä–∞–∑—Ä—è–¥—É –∏ –ø–æ–ª—É—á–∞–µ—Ç—Å—è 10)

**–ê—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏:**

def normalize_fruits(fruits: str) -> str:
    """
    –ù–æ—Ä–º–∞–ª–∏–∑—É–µ—Ç —Å—Ç—Ä–æ–∫—É —Å —Ñ—Ä—É–∫—Ç–∞–º–∏, –ø—Ä–∏–≤–æ–¥—è –µ—ë –∫ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º—É –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—é,
    –∏—Å–ø–æ–ª—å–∑—É—è –ø—Ä–∞–≤–∏–ª–∞ –æ–±–º–µ–Ω–∞ —Ñ—Ä—É–∫—Ç–æ–≤.

    Args:
        fruits: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏ (üçé, üçê, üçâ, üß∫).

    Returns:
        –°—Ç—Ä–æ–∫–∞ —Å –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Ñ—Ä—É–∫—Ç–æ–≤.
    """
    apples = fruits.count('üçé')
    pears = fruits.count('üçê')
    melons = fruits.count('üçâ')
    baskets = fruits.count('üß∫')

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —è–±–ª–æ–∫ –≤ –≥—Ä—É—à–∏
    pears += apples // 3
    apples %= 3

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≥—Ä—É—à –≤ –¥—ã–Ω–∏
    melons += (pears * 3) // 5
    pears %= 5

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –¥—ã–Ω—å –≤ –∫–æ—Ä–∑–∏–Ω—ã
    baskets += melons // 2
    melons %= 2

    # –°–æ–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫—É –æ–±—Ä–∞—Ç–Ω–æ, —Å–Ω–∞—á–∞–ª–∞ –∫–æ—Ä–∑–∏–Ω—ã, –ø–æ—Ç–æ–º –¥—ã–Ω–∏, –≥—Ä—É—à–∏, —è–±–ª–æ–∫–∏
    return (
        "üß∫" * baskets
        + "üçâ" * melons
        + "üçê" * pears
        + "üçé" * apples
    )


def add_fruits(fruits1: str, fruits2: str) -> str:
    """
    –°–∫–ª–∞–¥—ã–≤–∞–µ—Ç –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏.

    Args:
        fruits1: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏.
        fruits2: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏.

    Returns:
        –°—Ç—Ä–æ–∫–∞ —Å —Å—É–º–º–æ–π —Ñ—Ä—É–∫—Ç–æ–≤.
    """
    return normalize_fruits(fruits1 + fruits2)


def sub_fruits(fruits1: str, fruits2: str) -> str:
    """
    –í—ã—á–∏—Ç–∞–µ—Ç –≤—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É —Ñ—Ä—É–∫—Ç–æ–≤ –∏–∑ –ø–µ—Ä–≤–æ–π, –µ—Å–ª–∏ —ç—Ç–æ –≤–æ–∑–º–æ–∂–Ω–æ.

    Args:
        fruits1: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏, –∏–∑ –∫–æ—Ç–æ—Ä–æ–π –≤—ã—á–∏—Ç–∞–µ–º.
        fruits2: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏, –∫–æ—Ç–æ—Ä—É—é –≤—ã—á–∏—Ç–∞–µ–º.

    Returns:
        –°—Ç—Ä–æ–∫–∞ —Å —Ä–∞–∑–Ω–æ—Å—Ç—å—é —Ñ—Ä—É–∫—Ç–æ–≤ –∏–ª–∏ "–ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã—á–µ—Å—Ç—å", –µ—Å–ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π.
    """

    apples1 = fruits1.count('üçé')
    pears1 = fruits1.count('üçê')
    melons1 = fruits1.count('üçâ')
    baskets1 = fruits1.count('üß∫')

    apples2 = fruits2.count('üçé')
    pears2 = fruits2.count('üçê')
    melons2 = fruits2.count('üçâ')
    baskets2 = fruits2.count('üß∫')


    # –í—Ä–µ–º–µ–Ω–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –≤ –≤–∏–¥–µ –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —è–±–ª–æ–∫
    total_apples1 = apples1 + pears1 * 3 + melons1 * 15 // 3 + baskets1 * 30
    total_apples2 = apples2 + pears2 * 3 + melons2 * 15 // 3 + baskets2 * 30

    if total_apples1 < total_apples2:
        return "–ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã—á–µ—Å—Ç—å"
    else:
        total_apples = total_apples1 - total_apples2

    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —Å—É–º–º—ã —è–±–ª–æ–∫
    result_fruits = ""
    baskets = total_apples // 30
    result_fruits += "üß∫" * baskets
    total_apples %= 30
    melons = (total_apples*3) // 15
    result_fruits += "üçâ" * melons
    total_apples %= 15
    pears = total_apples // 3
    result_fruits += "üçê" * pears
    total_apples %= 3
    result_fruits += "üçé" * total_apples

    return normalize_fruits(result_fruits)

*   –ò—Å–ø–æ–ª—å–∑—É–π Python –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ —Å–≤–æ–∏—Ö —Ä–µ—à–µ–Ω–∏–π –∏ –∞–≤—Ç–æ–º–∞—Ç–∏–∑–∞—Ü–∏–∏ –ø–µ—Ä–µ–≤–æ–¥–∞.



````python
from typing import Union

def bin_to_dec(binary: str) -> int:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –¥–≤–æ–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Å—Ç—Ä–æ–∫—É) –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ.

    Args:
        binary: –î–≤–æ–∏—á–Ω–æ–µ —á–∏—Å–ª–æ –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏.

    Returns:
        –î–µ—Å—è—Ç–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ).
    """
    decimal = 0
    power = 0
    for digit in reversed(binary):
        if digit == '1':
            decimal += 2 ** power
        power += 1
    return decimal


def dec_to_bin(decimal: int) -> str:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Ü–µ–ª–æ–µ) –≤ –¥–≤–æ–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ (—Å—Ç—Ä–æ–∫—É).

    Args:
        decimal: –î–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ.

    Returns:
        –î–≤–æ–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Å—Ç—Ä–æ–∫–∞).
    """
    if decimal == 0:
        return "0"
    binary = ""
    while decimal > 0:
        binary = str(decimal % 2) + binary
        decimal = decimal // 2
    return binary


def ternary_to_dec(ternary: str) -> int:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Ç—Ä–æ–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Å—Ç—Ä–æ–∫—É) –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ.

    Args:
        ternary: –¢—Ä–æ–∏—á–Ω–æ–µ —á–∏—Å–ª–æ –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏.

    Returns:
        –î–µ—Å—è—Ç–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ).
    """
    decimal = 0
    power = 0
    for digit in reversed(ternary):
        decimal += int(digit) * (3 ** power)
        power += 1
    return decimal


def dec_to_ternary(decimal: int) -> str:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Ü–µ–ª–æ–µ) –≤ —Ç—Ä–æ–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ (—Å—Ç—Ä–æ–∫—É).

    Args:
        decimal: –î–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ.

    Returns:
        –¢—Ä–æ–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Å—Ç—Ä–æ–∫–∞).
    """
    if decimal == 0:
        return "0"
    ternary = ""
    while decimal > 0:
        ternary = str(decimal % 3) + ternary
        decimal = decimal // 3
    return ternary


def septenary_to_dec(septenary: str) -> int:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —Å–µ–º–µ—Ä–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Å—Ç—Ä–æ–∫—É) –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ.

    Args:
        septenary: –°–µ–º–µ—Ä–∏—á–Ω–æ–µ —á–∏—Å–ª–æ –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏.

    Returns:
        –î–µ—Å—è—Ç–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ).
    """
    decimal = 0
    power = 0
    for digit in reversed(septenary):
        decimal += int(digit) * (7 ** power)
        power += 1
    return decimal


def dec_to_septenary(decimal: int) -> str:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Ü–µ–ª–æ–µ) –≤ —Å–µ–º–µ—Ä–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ (—Å—Ç—Ä–æ–∫—É).

    Args:
        decimal: –î–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ.

    Returns:
        –°–µ–º–µ—Ä–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Å—Ç—Ä–æ–∫–∞).
    """
    if decimal == 0:
        return "0"
    septenary = ""
    while decimal > 0:
        septenary = str(decimal % 7) + septenary
        decimal = decimal // 7
    return septenary


def hex_to_dec(hexadecimal: str) -> int:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Å—Ç—Ä–æ–∫—É) –≤ –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ.

    Args:
        hexadecimal: –®–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–æ–µ —á–∏—Å–ª–æ –≤ –≤–∏–¥–µ —Å—Ç—Ä–æ–∫–∏.

    Returns:
        –î–µ—Å—è—Ç–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Ü–µ–ª–æ–µ —á–∏—Å–ª–æ).
    """
    decimal = 0
    power = 0
    for digit in reversed(hexadecimal):
        if digit.isdigit():
            decimal += int(digit) * (16 ** power)
        elif digit.upper() == 'A':
            decimal += 10 * (16 ** power)
        elif digit.upper() == 'B':
            decimal += 11 * (16 ** power)
        elif digit.upper() == 'C':
            decimal += 12 * (16 ** power)
        elif digit.upper() == 'D':
            decimal += 13 * (16 ** power)
        elif digit.upper() == 'E':
            decimal += 14 * (16 ** power)
        elif digit.upper() == 'F':
            decimal += 15 * (16 ** power)
        power += 1
    return decimal


def dec_to_hex(decimal: int) -> str:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –¥–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ (—Ü–µ–ª–æ–µ) –≤ —à–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ (—Å—Ç—Ä–æ–∫—É).

    Args:
        decimal: –î–µ—Å—è—Ç–∏—á–Ω–æ–µ —á–∏—Å–ª–æ.

    Returns:
        –®–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ (—Å—Ç—Ä–æ–∫–∞).
    """
    if decimal == 0:
        return "0"
    hex_digits = "0123456789ABCDEF"
    hexadecimal = ""
    while decimal > 0:
        remainder = decimal % 16
        hexadecimal = hex_digits[remainder] + hexadecimal
        decimal = decimal // 16
    return hexadecimal


def calculate_day_of_week(start_day: int, days_passed: float) -> int:
    """
    –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –ø–æ—Å–ª–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–Ω–µ–π.

    Args:
        start_day: –ù–∞—á–∞–ª—å–Ω—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ (0 - –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, 6 - –≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ).
        days_passed: –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—à–µ–¥—à–∏—Ö –¥–Ω–µ–π (–º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º).

    Returns:
        –î–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –ø–æ—Å–ª–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –¥–Ω–µ–π (0 - –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, 6 - –≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ).
    """
    if not isinstance(start_day, int) or not (0 <= start_day <= 6):
        raise ValueError("–ù–∞—á–∞–ª—å–Ω—ã–π –¥–µ–Ω—å –Ω–µ–¥–µ–ª–∏ –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Ü–µ–ª—ã–º —á–∏—Å–ª–æ–º –æ—Ç 0 –¥–æ 6 (–ø–Ω-–≤—Å)")
    if not isinstance(days_passed, (int, float)):
        raise ValueError("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ—à–µ–¥—à–∏—Ö –¥–Ω–µ–π –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —á–∏—Å–ª–æ–º")

    days_passed = int(days_passed)
    new_day = (start_day + days_passed) % 7
    return new_day


def day_number_to_name(day_number: int) -> str:
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç –Ω–æ–º–µ—Ä –¥–Ω—è –Ω–µ–¥–µ–ª–∏ (0-6) –≤ –µ–≥–æ –Ω–∞–∑–≤–∞–Ω–∏–µ.

    Args:
        day_number: –ù–æ–º–µ—Ä –¥–Ω—è –Ω–µ–¥–µ–ª–∏ (0 - –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, 6 - –≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ).

    Returns:
        –ù–∞–∑–≤–∞–Ω–∏–µ –¥–Ω—è –Ω–µ–¥–µ–ª–∏ (—Å—Ç—Ä–æ–∫–∞).
    """
    days = ["–ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫", "–≤—Ç–æ—Ä–Ω–∏–∫", "—Å—Ä–µ–¥–∞", "—á–µ—Ç–≤–µ—Ä–≥", "–ø—è—Ç–Ω–∏—Ü–∞", "—Å—É–±–±–æ—Ç–∞", "–≤–æ—Å–∫—Ä–µ—Å–µ–Ω—å–µ"]
    return days[day_number]


def normalize_fruits(fruits: str) -> str:
    """
    –ù–æ—Ä–º–∞–ª–∏–∑—É–µ—Ç —Å—Ç—Ä–æ–∫—É —Å —Ñ—Ä—É–∫—Ç–∞–º–∏, –ø—Ä–∏–≤–æ–¥—è –µ—ë –∫ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º—É –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—é,
    –∏—Å–ø–æ–ª—å–∑—É—è –ø—Ä–∞–≤–∏–ª–∞ –æ–±–º–µ–Ω–∞ —Ñ—Ä—É–∫—Ç–æ–≤.

    Args:
        fruits: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏ (üçé, üçê, üçâ, üß∫).

    Returns:
        –°—Ç—Ä–æ–∫–∞ —Å –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Ñ—Ä—É–∫—Ç–æ–≤.
    """
    if not all(fruit in ["üçé", "üçê", "üçâ", "üß∫"] for fruit in fruits):
        raise ValueError("–°—Ç—Ä–æ–∫–∞ –º–æ–∂–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ç–æ–ª—å–∫–æ —Å–∏–º–≤–æ–ª—ã üçé, üçê, üçâ, üß∫")

    apples = fruits.count('üçé')
    pears = fruits.count('üçê')
    melons = fruits.count('üçâ')
    baskets = fruits.count('üß∫')

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —è–±–ª–æ–∫ –≤ –≥—Ä—É—à–∏
    pears += apples // 3
    apples %= 3

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≥—Ä—É—à –≤ –¥—ã–Ω–∏
    melons += (pears * 3) // 5
    pears %= 5

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –¥—ã–Ω—å –≤ –∫–æ—Ä–∑–∏–Ω—ã
    baskets += melons // 2
    melons %= 2

    # –°–æ–±–∏—Ä–∞–µ–º —Å—Ç—Ä–æ–∫—É –æ–±—Ä–∞—Ç–Ω–æ, —Å–Ω–∞—á–∞–ª–∞ –∫–æ—Ä–∑–∏–Ω—ã, –ø–æ—Ç–æ–º –¥—ã–Ω–∏, –≥—Ä—É—à–∏, —è–±–ª–æ–∫–∏
    return (
        "üß∫" * baskets
        + "üçâ" * melons
        + "üçê" * pears
        + "üçé" * apples
    )


def add_fruits(fruits1: str, fruits2: str) -> str:
    """
    –°–∫–ª–∞–¥—ã–≤–∞–µ—Ç –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏.

    Args:
        fruits1: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏.
        fruits2: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏.

    Returns:
        –°—Ç—Ä–æ–∫–∞ —Å —Å—É–º–º–æ–π —Ñ—Ä—É–∫—Ç–æ–≤.
    """
    return normalize_fruits(fruits1 + fruits2)

def sub_fruits(fruits1: str, fruits2: str) -> str:
    """
    –í—ã—á–∏—Ç–∞–µ—Ç –≤—Ç–æ—Ä—É—é —Å—Ç—Ä–æ–∫—É —Ñ—Ä—É–∫—Ç–æ–≤ –∏–∑ –ø–µ—Ä–≤–æ–π, –µ—Å–ª–∏ —ç—Ç–æ –≤–æ–∑–º–æ–∂–Ω–æ.

    Args:
        fruits1: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏, –∏–∑ –∫–æ—Ç–æ—Ä–æ–π –≤—ã—á–∏—Ç–∞–µ–º.
        fruits2: –°—Ç—Ä–æ–∫–∞ —Å —Ñ—Ä—É–∫—Ç–∞–º–∏, –∫–æ—Ç–æ—Ä—É—é –≤—ã—á–∏—Ç–∞–µ–º.

    Returns:
        –°—Ç—Ä–æ–∫–∞ —Å —Ä–∞–∑–Ω–æ—Å—Ç—å—é —Ñ—Ä—É–∫—Ç–æ–≤ –∏–ª–∏ "–ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã—á–µ—Å—Ç—å", –µ—Å–ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–π.
    """
    if not all(fruit in ["üçé", "üçê", "üçâ", "üß∫"] for fruit in fruits1+fruits2):
        raise ValueError("–°—Ç—Ä–æ–∫–∞ –º–æ–∂–µ—Ç —Å–æ–¥–µ—Ä–∂–∞—Ç—å —Ç–æ–ª—å–∫–æ —Å–∏–º–≤–æ–ª—ã üçé, üçê, üçâ, üß∫")

    apples1 = fruits1.count('üçé')
    pears1 = fruits1.count('üçê')
    melons1 = fruits1.count('üçâ')
    baskets1 = fruits1.count('üß∫')

    apples2 = fruits2.count('üçé')
    pears2 = fruits2.count('üçê')
    melons2 = fruits2.count('üçâ')
    baskets2 = fruits2.count('üß∫')


    # –í—Ä–µ–º–µ–Ω–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –≤ –≤–∏–¥–µ –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —è–±–ª–æ–∫
    total_apples1 = apples1 + pears1 * 3 + melons1 * 15 // 3 + baskets1 * 30
    total_apples2 = apples2 + pears2 * 3 + melons2 * 15 // 3 + baskets2 * 30

    if total_apples1 < total_apples2:
        return "–ù–µ–≤–æ–∑–º–æ–∂–Ω–æ –≤—ã—á–µ—Å—Ç—å"
    else:
        total_apples = total_apples1 - total_apples2

    # –í–æ–∑–≤—Ä–∞—â–∞–µ–º –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ —Å—É–º–º—ã —è–±–ª–æ–∫
    result_fruits = ""
    baskets = total_apples // 30
    result_fruits += "üß∫" * baskets
    total_apples %= 30
    melons = (total_apples*3) // 15
    result_fruits += "üçâ" * melons
    total_apples %= 15
    pears = total_apples // 3
    result_fruits += "üçê" * pears
    total_apples %= 3
    result_fruits += "üçé" * total_apples

    return normalize_fruits(result_fruits)

# --- –ü—Ä–∏–º–µ—Ä—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è ---

print("--- –°–∏—Å—Ç–µ–º—ã —Å—á–∏—Å–ª–µ–Ω–∏—è ---")
binary_number = "1011"
decimal_number = bin_to_dec(binary_number)
print(f"–î–≤–æ–∏—á–Ω–æ–µ {binary_number} = –î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number}")

decimal_number = 11
binary_number = dec_to_bin(decimal_number)
print(f"–î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number} = –î–≤–æ–∏—á–Ω–æ–µ {binary_number}")

ternary_number = "210"
decimal_number = ternary_to_dec(ternary_number)
print(f"–¢—Ä–æ–∏—á–Ω–æ–µ {ternary_number} = –î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number}")

decimal_number = 21
ternary_number = dec_to_ternary(decimal_number)
print(f"–î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number} = –¢—Ä–æ–∏—á–Ω–æ–µ {ternary_number}")

septenary_number = "345"
decimal_number = septenary_to_dec(septenary_number)
print(f"–°–µ–º–µ—Ä–∏—á–Ω–æ–µ {septenary_number} = –î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number}")

decimal_number = 180
septenary_number = dec_to_septenary(decimal_number)
print(f"–î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number} = –°–µ–º–µ—Ä–∏—á–Ω–æ–µ {septenary_number}")

hex_number = "2AF"
decimal_number = hex_to_dec(hex_number)
print(f"–®–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–æ–µ {hex_number} = –î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number}")

decimal_number = 687
hex_number = dec_to_hex(decimal_number)
print(f"–î–µ—Å—è—Ç–∏—á–Ω–æ–µ {decimal_number} = –®–µ—Å—Ç–Ω–∞–¥—Ü–∞—Ç–µ—Ä–∏—á–Ω–æ–µ {hex_number}")

print("\n--- –î–Ω–∏ –Ω–µ–¥–µ–ª–∏ ---")
start_day = 0  # –ü–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫
days = 10.5 #–ü–æ–ª—Ç–æ—Ä—ã –Ω–µ–¥–µ–ª–∏
new_day = calculate_day_of_week(start_day, days)
print(f"{days} –¥–Ω—è –ø–æ—Å–ª–µ {day_number_to_name(start_day)}: {day_number_to_name(new_day)}")
days = 120 # –ß–µ—Ç—ã—Ä–µ –º–µ—Å—è—Ü–∞ (–ø—Ä–∏–±–ª–∏–∑–∏—Ç–µ–ª—å–Ω–æ)
new_day = calculate_day_of_week(start_day, days)
print(f"{days} –¥–Ω—è –ø–æ—Å–ª–µ {day_number_to_name(start_day)}: {day_number_to_name(new_day)}")

start_day = 4  # –ü—è—Ç–Ω–∏—Ü–∞
days = 365 # –ì–æ–¥
new_day = calculate_day_of_week(start_day, days)
print(f"{days} –¥–Ω—è –ø–æ—Å–ª–µ {day_number_to_name(start_day)}: {day_number_to_name(new_day)}")

print("\n--- –§—Ä—É–∫—Ç—ã ---")
fruits1 = "üçéüçéüçéüçéüçé" # 5 —è–±–ª–æ–∫
fruits2 = "üçéüçéüçé" # 3 —è–±–ª–æ–∫–∞
print(f"{fruits1} + {fruits2} = {add_fruits(fruits1, fruits2)}")

fruits3 = "üçêüçê"  # 2 –≥—Ä—É—à–∏
fruits4 = "üçéüçéüçéüçé" # 4 —è–±–ª–æ–∫–∞
print(f"{fruits3} + {fruits4} = {add_fruits(fruits3, fruits4)}")

fruits5 = "üçâüçâ" # 2 –¥—ã–Ω–∏
fruits6 = "üçéüçéüçéüçéüçéüçéüçéüçéüçéüçéüçéüçéüçéüçéüçé" # 15 —è–±–ª–æ–∫
print(f"{fruits5} + {fruits6} = {add_fruits(fruits5, fruits6)}")

fruits7 = "üß∫üß∫" # 2 –∫–æ—Ä–∑–∏–Ω—ã
fruits8 = "üçâüçâüçâ" # 3 –¥—ã–Ω–∏
print(f"{fruits7} + {fruits8} = {add_fruits(fruits7, fruits8)}")

fruits9 = "üß∫üçâüçêüçé" # 1 –∫–æ—Ä–∑–∏–Ω–∞, 1 –¥—ã–Ω—è, 1 –≥—Ä—É—à–∞, 1 —è–±–ª–æ–∫–æ
fruits10 = "üçâüçêüçé" # 1 –¥—ã–Ω—è, 1 –≥—Ä—É—à–∞, 1 —è–±–ª–æ–∫–æ
print(f"{fruits9} - {fruits10} = {sub_fruits(fruits9, fruits10)}")

fruits11 = "üß∫üçâ" # 1 –∫–æ—Ä–∑–∏–Ω–∞, 1 –¥—ã–Ω—è
fruits12 = "üß∫üçâüçéüçéüçé" # 1 –∫–æ—Ä–∑–∏–Ω–∞, 1 –¥—ã–Ω—è, 3 —è–±–ª–æ–∫–∞
print(f"{fruits11} - {fruits12} = {sub_fruits(fruits11, fruits12)}")

fruits13 = "üçâüçâüçâ" # 3 –¥—ã–Ω–∏
fruits14 = "üçéüçéüçéüçé" # 4 —è–±–ª–æ–∫–∞
print(f"{fruits13} - {fruits14} = {sub_fruits(fruits13, fruits14)}")

fruits15 = "üçêüçêüçêüçêüçê" # 5 –≥—Ä—É—à
fruits16 = "üçâ" # 1 –¥—ã–Ω—è
print(f"{fruits15} - {fruits16} = {sub_fruits(fruits15, fruits16)}")
````
